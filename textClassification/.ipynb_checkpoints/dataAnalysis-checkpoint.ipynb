```python
 # -*- coding:utf-8 -*-
import jieba
import pandas as pd
import numpy as np
import os
import xgboost
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

```python
df_data = pd.read_csv("./data/validData.csv")
print df_data
```

```python
df_data = pd.DataFrame(df_data, columns=["extract", "rel"])
print df_data
```

```python
print df_data["rel"].value_counts()
```

```python
corpus = df_data["extract"].tolist()
label = df_data["rel"].tolist()
```

```python
for string in corpus:
    print string, type(string)
```

```python
new_corpus = [x.decode("utf-8") for x in corpus]
label_mapping = {"-": 0,
                "撤县（市）设区": 1,
                "撤县设市": 2,
                "合并": 3,
                "新增县／市／区": 4,
                "撤地区设市": 5,
                "更名": 6,
                "归属变化": 7,
                "撤销": 8,
                "分解": 9}
new_label = [label_mapping[label_] for label_ in label]
```

```python
for string in new_corpus:
    print string, type(string)
```

```python
print new_label
```

```python
print len(new_corpus), len(new_label)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(new_corpus, new_label, test_size=0.2)
```

```python
print y_test
```

```python
from collections import Counter
print Counter(y_test)
```

```python
vectorizer = CountVectorizer()
tfidftransformer = TfidfTransformer()
tfidf = tfidftransformer.fit_transform(vectorizer.fit_transform(x_train))
weight = tfidf.toarray()
print tfidf.shape
test_tfidf = tfidftransformer.transform(vectorizer.transform(x_test))
test_weight = test_tfidf.toarray()
print test_weight.shape
```

```python
import xgboost as xgb
del bst
xgbtrain = xgb.DMatrix(weight, label=y_train)
xgbtest = xgb.DMatrix(test_weight, label=y_test)  # label可以不要，此处需要是为了测试效果
param = {'max_depth':6, 'eta':0.1, 'eval_metric':'merror', 'silent':1, 'objective':'multi:softmax', 'num_class':10}  # 参数
evallist  = [(xgbtrain,'train'), (xgbtest,'test')]  # 这步可以不要，用于测试效果
num_round = 50  # 循环次数
bst = xgb.train(param, xgbtrain, num_round, evallist)
preds = bst.predict(xgbtest)
```
